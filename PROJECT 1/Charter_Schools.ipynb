```python
# Dependencies and Setup
import csv 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
```

```python
charter_schools_data = pd.read_csv("1617_CS_ACHIEVEMENT.csv") 
charter_schools_data.head()
```

```python
charter_schools_df = charter_schools_data.loc[:, ['Building Name', 'District Name', 'County', 'City and Zip Code', 'Performance Index Percent 2016-17']]                                

```

```python
charter_schools_df[["City", "State", "Zip Code"]] = charter_schools_df["City and Zip Code"].str.split(",",expand=True).rename(columns=lambda x: f"string_{x+1}")
charter_schools_df.head()
```

```python
charter_schools_df[["Zip Code", "Code+4"]] = charter_schools_df["Zip Code"].str.split("-",expand=True).rename(columns=lambda x: f"string_{x+1}")
```

```python
charter_schools_df.columns
```

```python
charter_schools_df.drop(['City and Zip Code','Code+4'], axis=1)
```

```python
charter_schools_df = charter_schools_df.loc[charter_schools_df["Performance Index Percent 2016-17"] != "NC"]
```

```python
charter_schools_df.describe()
```

```python
charter_schools_franklin = charter_schools_df.loc[charter_schools_df["County"] == "Franklin"]
charter_schools_franklin.count()
```

```python
charter_schools_hamilton = charter_schools_df.loc[charter_schools_df["County"] == "Hamilton"]
charter_schools_hamilton.count()
```

```python
charter_schools_cuyahoga = charter_schools_df.loc[charter_schools_df["County"] == "Cuyahoga"]
charter_schools_cuyahoga.count()
```
